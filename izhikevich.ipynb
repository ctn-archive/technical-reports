```python
import nengo
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from nengo.neurons import Izhikevich
from nengo.dists import Uniform

model = nengo.Network()
with model:
    ens = nengo.Ensemble(30, dimensions=1, neuron_type=Izhikevich(), max_rates=Uniform(50, 250))
    nengo.Connection(nengo.Node(lambda t: np.sin(4 * np.pi * t)), ens)
    ep = nengo.Probe(ens, synapse=0.01)
    sp = nengo.Probe(ens.neurons)
    vp = nengo.Probe(ens.neurons, 'voltage')
    rp = nengo.Probe(ens.neurons, 'recovery')
sim = nengo.Simulator(model, dt=0.0005)
sim.run(5.0)
```

```python
from nengo.utils.ensemble import tuning_curves
eval_points, activities = tuning_curves(ens, sim)
plt.plot(eval_points, activities);
```
